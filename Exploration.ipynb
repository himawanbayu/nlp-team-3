```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cohort_plots/bert-base-uncased/bert_extrinsic_scores.csv")
summary = df.groupby("source")[["accuracy", "difficulty"]].mean().reset_index()
pivot_summary = summary.set_index("source")

# Compute difference per question
pivot_df = df.pivot(index="question", columns="source", values="accuracy").dropna()
pivot_df["accuracy_diff"] = pivot_df["sys"] - pivot_df["ref"]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

pivot_summary["accuracy"].plot(kind="bar", ax=axes[0], color=["lightcoral", "lightgreen"])
axes[0].set_title("Average Accuracy")
axes[0].set_ylabel("Accuracy")
axes[0].set_ylim(0, 1)

pivot_summary["difficulty"].plot(kind="bar", ax=axes[1], color=["lightcoral", "lightgreen"])
axes[1].set_title("Average Difficulty")
axes[1].set_ylabel("Difficulty")
axes[1].set_ylim(0, 1)

plt.tight_layout()
# plt.savefig("cohort_plots/bert-base-uncased/accuracy_difficulty_comparison.png")
plt.show()

# Summary
summary["metric"] = ["accuracy", "difficulty"]
summary_csv = summary.set_index("metric")[["accuracy", "difficulty"]]
summary_csv.to_csv("cohort_plots/bert-base-uncased/bert_extrinsic_summary.csv")
```
