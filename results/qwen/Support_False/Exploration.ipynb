albert-base-v2

```python
df = pd.read_csv("albert-base-v2/0.1/test_taker_results.csv")
summary = df.groupby("source")[["accuracy", "confidence"]].mean().reset_index()
pivot_summary = summary.set_index("source")

# Compute difference per question
pivot_df = df.pivot(index="question", columns="source", values="accuracy").dropna()
pivot_df["accuracy_diff"] = pivot_df["sys"] - pivot_df["ref"]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

colors = {"ref": "lightcoral", "sys": "lightgreen"}

# Accuracy bar plot
pivot_summary["accuracy"].plot(kind="bar", ax=axes[0],
                                color=[colors[src] for src in pivot_summary.index])
axes[0].set_title("Average Accuracy")
axes[0].set_ylabel("Accuracy")
axes[0].set_ylim(0, 1)
axes[0].set_xticklabels(pivot_summary.index, rotation=0)
for i, v in enumerate(pivot_summary["accuracy"]):
    axes[0].text(i, v + 0.02, f"{v:.2f}", ha="center", va="bottom", fontsize=9)
# Confidence
pivot_summary["confidence"].plot(kind="bar", ax=axes[1],
                                 color=[colors[src] for src in pivot_summary.index])
axes[1].set_title("Average Confidence")
axes[1].set_ylabel("Confidence")
axes[1].set_ylim(0, 1)
axes[1].set_xticklabels(pivot_summary.index, rotation=0)
for i, v in enumerate(pivot_summary["confidence"]):
    axes[1].text(i, v + 0.02, f"{v:.2f}", ha="center", va="bottom", fontsize=9)
plt.tight_layout()
plt.show()

# Summary
summary["metric"] = ["accuracy", "confidence"]
summary_csv = summary.set_index("source")[["accuracy", "confidence"]]
# summary_csv.to_csv("albert-base-v2/0.1/extrinsic_summary.csv")
```

```python
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

paths = glob.glob("*/0.1/test_taker_results.csv")

dfs = []
for path in paths:
    try:
        model_name = path.split("/")[0]  # just the model name
        df = pd.read_csv(path)
        df['model'] = model_name
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {path}: {e}")

if dfs:
    all_results_df = pd.concat(dfs, ignore_index=True)
    # all_results_df.to_csv("all_models_combined.csv", index=False)

    agg = all_results_df.groupby(['model', 'source'])['accuracy'].mean().reset_index()

    plt.figure(figsize=(10, 5))
    ax = sns.barplot(data=agg, x='model', y='accuracy', hue='source')

    # Add score labels on top of bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%.2f', label_type='edge', padding=2)

    plt.title("Accuracy per Model and Distractor Source")
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()
else:
    print("No dataframes loaded. Check if CSV files exist in the expected folder structure.")
```

Albert-base

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from collections import Counter
from sklearn.metrics import mean_squared_error, accuracy_score
from scipy.stats import pearsonr, spearmanr


df = pd.read_csv("albert-base-v2/0.1/test_taker_results.csv")

df['confidence'] = df['confidence'].astype(float)
df['accuracy'] = df['accuracy'].astype(int)

df['distractor_effectiveness'] = df['distractor_effectiveness'].apply(ast.literal_eval)
df['raw_preds'] = df['raw_preds'].apply(ast.literal_eval)

plt.figure(figsize=(8, 5))
for source in ['ref', 'sys']:
    sns.histplot(df[df['source'] == source]['confidence'], kde=False, label=source, bins=10, alpha=0.6)
plt.xlabel("Confidence (Correct prediction frequency)")
plt.ylabel("Number of Questions")
plt.title("Confidence Distribution by Distractor Source - ALBERT-base-v2")
plt.legend()
plt.show()

def get_top_wrong_choices(df, source='ref'):
    counter = Counter()
    for _, row in df[df['source'] == source].iterrows():
        choices = [row['correct_answer'], row['distractor1'], row['distractor2'], row['distractor3']]
        for pred in row['raw_preds']:
            if pred != 0:  # 0 = correct
                picked = choices[pred]
                counter[picked] += 1
    return counter.most_common(10)

ref_wrong = get_top_wrong_choices(df, 'ref')
sys_wrong = get_top_wrong_choices(df, 'sys')

def plot_top_distractors(wrong_counts, title):
    labels, values = zip(*wrong_counts)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(values), y=list(labels), palette='rocket')
    plt.title(title)
    plt.xlabel("Times Chosen by Model")
    plt.ylabel("Distractor")
    plt.show()

plot_top_distractors(ref_wrong, "Top 10 Most Selected Distractors (ref) - ALBERT-base-v2")
plot_top_distractors(sys_wrong, "Top 10 Most Selected Distractors (sys) - ALBERT-base-v2")

pearson_corr, _ = pearsonr(df['confidence'], df['accuracy'])
spearman_corr, _ = spearmanr(df['confidence'], df['accuracy'])

print(f"Pearson correlation between confidence and accuracy: {pearson_corr:.3f}")
print(f"Spearman correlation between confidence and accuracy: {spearman_corr:.3f}")

print("\nHigh confidence correct predictions:")
print(df[(df['confidence'] == 1.0) & (df['accuracy'] == 1)].head(3)[['question', 'confidence', 'accuracy', 'source']])

print("\nLow confidence incorrect predictions:")
print(df[(df['confidence'] < 0.4) & (df['accuracy'] == 0)].head(3)[['question', 'confidence', 'accuracy', 'source']])
```
